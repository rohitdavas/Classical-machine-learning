# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Data

## create data

```python
def makeData(N, err = 1.0,rseed = 1):
    rng = np.random.RandomState(rseed)
    
    x = rng.rand(N,1)**2
    y = 10 - 1/(x.ravel() + 0.1)
    
    if err > 0:
        y += err*rng.randn(N)
    return x, y
```

```python
x, y = makeData(100)

print(x.shape, y.shape)
```

## visualise data

```python
def plot(x,y):
    plt.scatter(x,y, c = 'g')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
plot(x,y)
```

## train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y)

print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

# Polynomial regression

## Imports

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
def polynomialRegression(degree = 2,**kwargs):
    return make_pipeline(PolynomialFeatures(degree), LinearRegression(**kwargs)) 
```

## model: Polynomial regression

```python
degrees = [1, 3, 5, 7, 9]

plt.figure(figsize=(10, 10))
plt.scatter(X_train, Y_train)

X_test2 = np.linspace(-0.1, 1.1, 500)[:, None]
for degree in degrees:
    Y_pred = polynomialRegression(degree).fit(X_train, Y_train).predict(X_test2)
    plt.plot(X_test2, Y_pred, label='degree{0}'.format(degree))

plt.legend(loc='best')
plt.show()
```

# validation curve

```python
from sklearn.model_selection import validation_curve
```

## define the validation curve

```python
degree = np.arange(0, 21)

train_score, val_score = validation_curve(polynomialRegression(),
                                          x,
                                          y,
                                          'polynomialfeatures__degree',
                                          degree,
                                          cv=3 #k-folds. for normal validation train folds = 2
                                         )
```

## get the median train and test score across the k-folds

```python
median_train_score = np.median(train_score, 1) 
median_test_score = np.median(val_score, 1) 
```

## visualise

```python
plt.plot(degree[:, np.newaxis], median_train_score, color = 'blue', label = 'training score')
plt.plot(degree[:,np.newaxis], median_test_score, color = 'red', label = 'test score')

plt.xlabel("degree")
plt.ylabel("score")
plt.ylim(0,1)
plt.legend()
plt.show()
```

## use the best degree from visualisation

```python
best_degree = 3 

plt.scatter(x,y)
y_test = polynomialRegression(best_degree).fit(x,y).predict(X_test2)
plt.plot(X_test2, y_test)
plt.show()
```
